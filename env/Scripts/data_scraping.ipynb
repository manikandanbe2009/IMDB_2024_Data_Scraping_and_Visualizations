```python
# 1. Genres Action
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import csv
from selenium.common.exceptions import NoSuchElementException, ElementClickInterceptedException
from selenium.webdriver.common.action_chains import ActionChains
driver = webdriver.Chrome()
driver.get("https://www.imdb.com/search/title/?title_type=feature&release_date=2024-01-01,2024-12-31&genres=action");
csv_file = open("action_genres.csv",mode='w',newline="",encoding="utf-8")
write = csv.writer(csv_file)
write.writerow(["Movie Name","Genre","Ratings","Voting Counts","Duration"])
load_count = 0
max_loads = 12

while load_count < max_loads:
    time.sleep(2)
    try:       

        button = driver.find_element(By.XPATH, '//*[@id="__next"]/main/div[2]/div[3]/section/section/div/section/section/div[2]/div/section/div[2]/div[2]/div[2]/div/span/button')
        print(button.text)
        ActionChains(driver).move_to_element(button).perform()
        button.click()
        print(f"Clicked '50 More' ({load_count + 1})")
        load_count += 1
        time.sleep(3)
    except (NoSuchElementException, ElementClickInterceptedException):
        print("No more Load More button or cannot click.")
        break

print(f"Finished loading. Total loads: {load_count}")

ul_elements = driver.find_elements(By.CSS_SELECTOR,'ul[class="ipc-metadata-list ipc-metadata-list--dividers-between sc-e22973a9-0 khSCXM detailed-list-view ipc-metadata-list--base"]');
for ul in ul_elements:
    li_elements = ul.find_elements(By.TAG_NAME, 'li')
    for li in li_elements:
        try:
            
            title_element = li.find_element(By.CSS_SELECTOR, "h3[class='ipc-title__text']")
            title =  title_element.text.replace(".", "", 1).split(" ", 1)[-1]
            duration_element = li.find_element(By.CSS_SELECTOR, "span:nth-of-type(2)[class='sc-2bbfc9e9-7 jttFlJ dli-title-metadata-item']")
            rating_element = li.find_element(By.CSS_SELECTOR, "span[class='ipc-rating-star--rating']")
            voting_element = li.find_element(By.CSS_SELECTOR, "span[class='ipc-rating-star--voteCount']")
            voting_elemet1 = voting_element.text.replace("(", "")
            voting = voting_elemet1.replace(")","")
            write.writerow([title,'Action',rating_element.text,voting,duration_element.text])
        except:
            
            print("No Data")
            continue
csv_file.close()
```

```python
import pandas as pd
actionGenre = pd.read_csv("action_genres.csv")
adventureGenres = pd.read_csv("adventure_genres.csv")
comedyGenres = pd.read_csv("comedy_genres.csv")
crimeGenres = pd.read_csv("crime_genres.csv")
thrillerGenres = pd.read_csv("thriller_genres.csv")
IMDBCombinedData = pd.concat([actionGenre,adventureGenres,comedyGenres,crimeGenres,thrillerGenres], axis=0)

IMDBCombinedData.info()
```

```python
def convert_to_number(value):
    if isinstance(value, str):
        value = value.strip().upper().replace(",", "")
        if value.endswith('K'):
            return float(value[:-1]) * 1000
        elif value.endswith('M'):
            return float(value[:-1]) * 1000000
        elif value.isdigit():
            return int(value)
    return 0

# Apply conversion once to the whole column
IMDBCombinedData['Voting Counts'] = IMDBCombinedData['Voting Counts'].apply(convert_to_number).astype(int)

# Preview
IMDBCombinedData.head()
```
